```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlretrieve

# Download sample image from the internet
image_url = 'https://upload.wikimedia.org/wikipedia/commons/7/7d/Dog_face.png'  # Example image URL
urlretrieve(image_url, 'downloaded_image.png')

# Load grayscale image
image = cv2.imread('downloaded_image.png', cv2.IMREAD_GRAYSCALE)

# Define Sobel filters explicitly as given
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

# Apply Sobel filters using cv2.filter2D
sobel_x_img = cv2.filter2D(image, -1, sobel_x)
sobel_y_img = cv2.filter2D(image, -1, sobel_y)

# Display images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title("Sobel-X")
plt.imshow(sobel_x_img, cmap='gray')
plt.subplot(1, 3, 3)
plt.title("Sobel-Y")
plt.imshow(sobel_y_img, cmap='gray')
plt.show()
```
